```python
import pandas
import numpy
from matplotlib import pyplot
```

```python
data = pandas.read_csv('train.csv')

data.head(892)
```

Przypisuję wartość do wieku

```python
data_copy =  data.copy(deep=True)


data_copy["age_group"] = pandas.cut(x=data_copy['Age'], bins=[0,30,60,100], labels=["young","middle_aged","old"])

data_copy.head()
```

Wyprowadzam wiek w kolejności rosnącej

```python
data_copy = data_copy.sort_values('Age', ascending=True)
data1 =  data_copy.copy(deep=True)

data1
```

Sortuję według ocalałych. Wniosek najpierw kobieta potem mężczyzna

```python
data1 = data1.sort_values(['Survived', 'Sex'], ascending= [False, 'famale'])

data1.head(892)
```

Przypisywanie wartości "przeżył lub umarł"

```python
data1["Survived_group"] = pandas.cut(x=data1['Survived'], bins=[-1, 0, 1,], labels=["Died", "Survived"])

data1.head(892)
```
